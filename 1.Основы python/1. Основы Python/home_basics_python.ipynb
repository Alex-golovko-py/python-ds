```python
phrase_1 = ['Насколько проще было бы писать программы, если бы не заказчики','641Кб должно хватить для любых задач. Билл Гейтс (по легенде)',
            'Насколько проще было бы писать программы, если бы не заказчики']
phrase_2 = ['640Кб должно хватить для любых задач. Билл Гейтс (по легенде)','Насколько проще было бы писать программы, если бы не заказчики',
            'Насколько проще было бы писать программы, если бы не заказчики']
```

```python
print(f'Длина первой строки: {len(phrase_1[0])}')
print(f'Длина второй строки: {len(phrase_2[0])}')

if len(phrase_1[0]) > len(phrase_2[0]):
    print(f'Длина первой строки больше второй на - {len(phrase_1[0]) - len(phrase_2[0])}')
elif len(phrase_1[0]) < len(phrase_2[0]):
    print(f'Длина первой строки больше второй на: {len(phrase_2[0]) - len(phrase_1[0])}')
else:
    print('Строки равны между собой')
```

<h2>Задание 2</h2>
Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы: 1.

year = 2020
Результат:
Високосный год

year = 2019
Результат:
Обычный год

```python
try:
    year=int(input('Укажите год'))
    if (year % 4 == 0 and year % 100 !=0):
        print(f'Год {year} является високосным')
    else:
        print(f'{year} не является високосным')
except ValueError:
    print(f' {year} - Не является числом')
```
